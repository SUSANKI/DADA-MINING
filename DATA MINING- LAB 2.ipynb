#### Problem 1.4
*Manipulating image with linear algebra

```python
from scipy import misc
from numpy import linalg
import matplotlib.pyplot as plt
img = misc.face()
plt.imshow(img)
plt.show()
```

 Get shape, axis and array properties

```python
img.shape # the shape property of image
img.ndim # the dimention number of image
img[:, :, 0]# the pixel matrix of image
```

 Manipulate on array

```python
img_array = img / 255
img_array.max(), img_array.min()
# multiplication operator @ to convert gray scale
img_gray = img_array @ [0.2126, 0.7152, 0.0722]
plt.imshow(img_gray, cmap="gray")
plt.show()
```

 Apply linear algebra on array (svd can be intensive computation)

```python
U, s, Vt = linalg.svd(img_gray) # take time
U.shape, s.shape, Vt.shape
```
